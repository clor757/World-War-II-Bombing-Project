```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csvfile = "C:/Users/bryan_j7mwyoj/Desktop/Project 2/operations.csv"
WWII_ops = pd.read_csv(csvfile, low_memory = False)
WWII_ops.head()
```

```python
WWII_ops.drop(['Callsign', 'Source ID', 'Target Type', 'Target Industry', 'Target Priority', 'Bomb Damage Assessment', 'Air Force',
              'Unit ID', 'Mission Type', 'Target ID', 'Aircraft Returned', 'Aircraft Failed', 'Aircraft Damaged', 'Aircraft Lost', 'Airborne Aircraft',
              'Attacking Aircraft', 'Bombing Aircraft', 'Time Over Target'], axis = 1, inplace = True)
WWII_ops.rename(columns = {'Altitude (Hundreds of Feet)': 'Altitude', 'High Explosives Weight (Pounds)': "High Explosives Weight LBS",
                          'High Explosives Weight (Tons)': 'High Explosives Weight Tons', 'Incendiary Devices Weight (Pounds)': 'Incendiary Devices Weight LBS',
                          'Incendiary Devices Weight (Tons)': 'Incendiary Devices Weight Tons', 'Fragmentation Devices Weight (Pounds)': 'Fragmentation Devices Weight LBS',
                          'Fragmentation Devices Weight (Tons)': 'Fragmentation Devices Weight Tons',
                          'Total Weight (Pounds)': 'Total Weight LBS', 'Total Weight (Tons)': 'Total Weight Tons'}, inplace = True)
```

```python
WWII_ops
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/WWII_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
WWII_ops.to_sql(name = 'WWII_ops', con=engine, if_exists='append', index=False)
```
