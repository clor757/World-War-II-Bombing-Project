```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
csvfile = "C:/Users/bryan_j7mwyoj/Desktop/Project 2/World-War-II-Bombing-Project/operations.csv"
WWII_ops = pd.read_csv(csvfile, low_memory = False)
WWII_ops.head()
```

```python
WWII_ops.drop(['Callsign', 'Source ID', 'Target Type', 'Target Industry', 'Target Priority', 'Bomb Damage Assessment', 'Air Force',
              'Unit ID', 'Mission Type', 'Target ID', 'Aircraft Returned', 'Aircraft Failed', 'Aircraft Damaged', 'Aircraft Lost', 'Airborne Aircraft',
              'Attacking Aircraft', 'Bombing Aircraft', 'Time Over Target'], axis = 1, inplace = True)
```

```python
WWII_ops.rename(columns = {'Altitude (Hundreds of Feet)': 'Altitude', 'High Explosives Weight (Pounds)': "High_Explosives_Weight_LBS",
                          'High Explosives Weight (Tons)': 'High_Explosives_Weight_Tons', 'Incendiary Devices Weight (Pounds)': 'Incendiary_Devices_Weight_LBS',
                          'Incendiary Devices Weight (Tons)': 'Incendiary_Devices_Weight_Tons', 'Fragmentation Devices Weight (Pounds)': 'Fragmentation_Devices_Weight_LBS',
                          'Fragmentation Devices Weight (Tons)': 'Fragmentation_Devices_Weight_Tons',
                          'Total Weight (Pounds)': 'Total_Weight_LBS', 'Total Weight (Tons)': 'Total_Weight_Tons',
                          'Mission ID':'Mission_ID', 'Mission Date':'Mission_Date', 'Aircraft Series':'Aircraft_Series',
                          'Takeoff Base':'Takeoff_Base', 'Takeoff Location':'Takeoff_Location', 'Takeoff Latitude':'Takeoff_Latitude',
                          'Takeoff Longitude':'Takeoff_Longitude', 'Target Country':'Target_Country',
                          'Target City':'Target_City', 'Target Latitude':'Target_Latitude', 'Target Longitude':'Target_Longitude', 'Theater of Operations':'Theater_of_Operations'}, inplace = True)
```

```python
WWII_ops
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/WWII_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
WWII_ops.to_sql(name = 'WWII_ops', con=engine, if_exists='append', index=False)
```

```python
WWII_ops.to_csv('WWII_ops.csv')
```

```python
WWII_ops2=WWII_ops.drop_duplicates(['Takeoff_Latitude', 'Takeoff_Longitude', 'Target_Latitude', 'Target_Longitude'], keep='last')
```

```python
WWII_ops2
```

```python
WWIIops2=WWII_ops2.dropna(subset= ['Takeoff_Latitude', 'Takeoff_Longitude', 'Target_Latitude', 'Target_Longitude'])
WWIIops2
```

```python
WWIIops2.to_sql(name = 'WWIIops2', con=engine, if_exists='append', index=False)
```

```python
WWIIops2.to_csv('WWIIops2.csv')
```
